```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path
import warnings
warnings.filterwarnings("ignore")
```

```python
# File to Load 
zhvi_zip_load = Path("data files\Zip_zhvi_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv")

# Read data files and store into Pandas DataFrames
df_zhvi_zip = pd.read_csv(zhvi_zip_load)

df_zhvi_zip.head()

# # Combine the data into a single dataset.  
# school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
# school_data_complete.head()
```

```python
new_df_zhvi_zip = df_zhvi_zip[['SizeRank', 'RegionName', 'State', 'City', '1/31/2018', '1/31/2019', '1/31/2020', '1/31/2021', '1/31/2022', '1/31/2023']]
new_df_zhvi_zip = new_df_zhvi_zip[new_df_zhvi_zip["State"].str.contains("TX")]
new_df_zhvi_zip.head()
```

```python
new_df_zhvi_zip.rename(columns = {'RegionName':'ZipCode'}, inplace = True)
new_df_zhvi_zip.columns
```

```python
new_df_zhvi_zip.isnull()
```

```python
new_df_zhvi_zip = new_df_zhvi_zip.dropna()


new_df_zhvi_zip.isnull()
```

```python
import numpy as np
new_df_zhvi_zip['12/31/2023'] = np.nan
new_df_zhvi_zip
```

```python
# File to Load 
predictions_load = Path("data files\predictions.csv")

# Read data files and store into Pandas DataFrames
df_predictions = pd.read_csv(predictions_load)

df_predictions.head()

# Combine the data into a single dataset.  
zip_zhvi_complete = pd.concat([new_df_zhvi_zip, df_predictions], axis=0, ignore_index=True)
zip_zhvi_complete
```

```python
zip_zhvi_complete.to_csv('data files\zipZHVI_main.csv')
```
